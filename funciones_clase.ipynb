```python
import numpy as np
import random

def crear_tablero(lado = 10):
    tablero = np.full((lado, lado), ' ')
    return tablero


def coloca_barco(tablero, barco):
    for pieza in barco:
        tablero[pieza] = 'O' # Esto con una lista fallaría
    return tablero


def coloca_barco_plus(tablero, barco):
    # Devuelve el tablero si puede colocar el barco, si no, devuelve False y avisa por pantalla
    # VERIFICAR QUE NO HAYA UN BARCO A TROZOS Y QUE LAS COORDENADAS SON CONSECUTIVAS
    
    tablero_temp = tablero.copy()
    num_max_filas = tablero.shape[0]
    num_max_columnas = tablero.shape[1]
    
    for pieza in barco:
        fila = pieza[0]
        columna = pieza[1]
        
        if fila < 0 or fila >= num_max_filas:
            print(f'No puedo poner la pieza {pieza} porque se sale del tablero')
            return False
        
        if columna < 0 or fila >= num_max_columnas:
            print(f'No puedo poner la pieza {pieza} porque se sale del tablero')
            return False
        
        if tablero[pieza] == 'O' or tablero[pieza] == 'X':
            print(f'No puedo poner la pieza {pieza} porque ya hay otro barco')
            return False
        
        tablero_temp[pieza] = 'O'
        
    return tablero_temp # se puede sobreescribir de tablero original también


def crea_barco_aleatorio(tablero, eslora = 4, num_intentos = 100):
    num_max_filas = tablero.shape[0]
    num_max_columnas = tablero.shape[1]
    
    while True:
        # Construimos el barco hipotético
        barco = []
        
        # Colocación de la primera pieza del barco
        pieza_original = (random.randint(0,num_max_filas-1), random.randint(0,num_max_columnas-1))
        print('Pieza original:', pieza_original)
        barco.append(pieza_original)
        
        # Colocación del resto de piezas
        orientacion = random.choice(['N', 'S', 'E', 'O'])
        print('Con orientación:', orientacion)
        
        fila = pieza_original[0]
        columna = pieza_original[1]
        
        for i in range(eslora - 1): # porque una de las piezas ya está puesta, por eso se resta 1
            # Se puede hacer con slicing comprobando que cabe la eslora completa --> len(slicing) = eslora + no incluye 'O' o 'X'
            if orientacion == 'N':
                fila -= 1 # va subiendo un por cada vuelta del bucle
            elif orientacion == 'S':
                fila += 1
            elif orientacion == 'E':
                columna += 1
            else:
                columna -= 1
                
            pieza = (fila, columna)
            barco.append(pieza)

        tablero_temp = coloca_barco_plus(tablero, barco)
        if type(tablero_temp) == np.ndarray:
            return tablero_temp
        else: 
            num_intentos -= 1
            print('Tengo que intentar colocar otro barco')
    
      
def recibir_disparo(tablero, coordenada):
    if tablero[coordenada] == 'O':
        tablero[coordenada] = 'X'
    elif tablero[coordenada] == 'X':
        # Para la máquina debe comprobar que no dispara 2 veces al mismo sitio
        print('¡Ya habías disparado ahí! La próxima vez dispara mejor a otro sitio')
    else:
        tablero[coordenada] = '-'
        print('¡Agua!')
    
    return tablero


def arregla_barcos(tablero):
    tablero[tablero == 'X'] = 'O'
    # tablero.where[tablero == 'X', 'O', tablero]
    return tablero
    
```

```python
tablero = crear_tablero()
barco1 = [(0,1), (1,1)]
barco2 = [(1,3), (1,4), (1,5), (1,6)]
    
print('Crear tablero:', tablero, sep = '\n')
for barco in [barco1, barco2]:
    print('')
    print('Coloco:', barco)
    tablero = coloca_barco(tablero, barco)
    print(tablero)
    
tablero = recibir_disparo(tablero, (5,6))
tablero = recibir_disparo(tablero, (1,1))
print('')
print(tablero)
    
tablero = arregla_barcos(tablero)
print('')
print(tablero)
```
